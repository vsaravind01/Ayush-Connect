```python
directory_path = "/content/MepcoTropicLeaf-V1/Database"
files = os.listdir(directory_path)
```

```python
plant_names = [os.path.splitext(file)[0] for file in files if file.endswith(".zip")]
```

```python
import requests
import pandas as pd


api_url = "https://api.gbif.org/v1/species/search?q="

plant_data_list = []

for plant_name in plant_names:
    query_url = api_url + plant_name.replace(" ", "%20")
    response = requests.get(query_url)

    if response.status_code == 200:
        plant_data = response.json()

        if "results" in plant_data and len(plant_data["results"]) > 0:
            result = plant_data["results"][0]
            data = {
                "Plant Name": plant_name,
                "Scientific Name": result.get("scientificName", "N/A"),
                "Canonical Name": result.get("canonicalName", "N/A"),
                "Parent": result.get("parent", "N/A"),
                "Kingdom": result.get("kingdom", "N/A"),
                "Order": result.get("order", "N/A"),
                "Family": result.get("family", "N/A"),
                "Genus": result.get("genus", "N/A"),
                "Species": result.get("species", "N/A"),
                "Class": result.get("class", "N/A"),
            }
            plant_data_list.append(data)
        else:
            print(f"No data found for '{plant_name}'")

df = pd.DataFrame(plant_data_list)
```
